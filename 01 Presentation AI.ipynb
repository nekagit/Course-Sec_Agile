# How to prompt

1. Be Specific and Clear
Include relevant context
State your desired outcome
Specify format or style requirements

2. Use System Instructions
Set the tone (formal/casual)
Define the role you want ChatGPT to take
Specify constraints or limitations

3. Break Down Complex Requests
Ask one thing at a time
Build upon previous responses
Iterate and refine your queries

"Act as a [role] and help me with [task]"
"I need to [goal]. Can you help me [specific request]?"
"Please provide [output] in [format] style"

### Tips for Success
Document effective prompts
Start with clear objectives
- Use iterative refinement
- Save effective prompts for future use
Combine ChatGPT with human expertise
Regular practice and experimentation

# History of ChatBots


## Alan Turing 1950s
- Turing Test
- Definition of AI
- 3 Rooms ( 1 Computer, 1 Abfragen, 1 Mensch )

## Early Foundations (1960s-1970s)

### ELIZA (1966)
- Created by Joseph Weizenbaum at MIT
- First natural language processing computer program
- Simulated a psychotherapist using pattern matching
- Used simple scripts to reformulate user input into questions
- Demonstrated the "ELIZA Effect" where humans attributed understanding to the program


### PARRY (1972)
- Developed by psychiatrist Kenneth Colby
- Simulated a person with paranoid schizophrenia
- More sophisticated than ELIZA with a system of beliefs and fears
- First chatbot to pass a version of the Turing test


### A.L.I.C.E. (1995)
- Artificial Linguistic Internet Computer Entity
- Created by Richard Wallace
- Used AIML (Artificial Intelligence Markup Language)
- Won multiple Loebner Prize competitions
- Influenced many subsequent chatbots


### IBM's Watson (2006)
- Advanced natural language processing
- Famous for winning Jeopardy! in 2011
- Later developed into business applications
- Demonstrated potential of AI in understanding context


### Siri (2010)
- Released by Apple
- First modern digital assistant on smartphones
- Combined voice recognition with natural language processing
- Set the standard for mobile AI assistants


### Microsoft's Cortana (2014)
- Integrated with Windows devices
- Named after Halo video game character
- Combined search capabilities with personal assistance
- Featured natural language interaction


### Amazon's Alexa (2014)
- Voice-activated assistant
- Integrated with smart home devices
- Expanded to third-party skills
- Popularized voice-based interaction


### Replika (2017)
- Personal AI companion
- Focused on emotional support
- Learns from conversations
- Uses advanced neural networks


### ChatGPT (2022)
- Created by OpenAI
- Based on GPT architecture
- Achieved widespread public adoption
- Demonstrated advanced reasoning capabilities
