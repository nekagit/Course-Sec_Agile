### Data security: Transfer : WhatsApp
- CIA Triad: Stands for Confidentiality, Integrity, and Availability
    - Confidentiality : Ensuring that sensitive information is not accessed by unauthorized individuals
        1. Multi-Factor Authentication (MFA) : Adds extra layers of security, requiring more than just a password to gain access, increasing confidentiality
        2. Encryption: Scrambles data so only authorized individuals can decode it, increasing confidentiality
    - Integrity : Ensuring the data received is the same as the data sent
        1. Hashing: The sender sends hash with data: (Data - > Hash X). The receiver decodes Hash X and compares if its equal to data, increasing integrity
        2. Certificates: Used to confirm the identify of devices or user and ensure data is not tampered with, increasing integrity
    - Availability : Ensuring that systems and data are accessible to authorized individuals when needed
        - Fault Tolerance: Backup
        - System Patching and Updates


### Authentication & Authorization & Accounting

| **Concept**                      | **Definition**                                                                                  | **Example**                                                                                 | **Purpose**                                                                                               |
|-----------------------------------|-------------------------------------------------------------------------------------------------|---------------------------------------------------------------------------------------------|-----------------------------------------------------------------------------------------------------------|
| **Identification**                | Claiming a specific identity on a system.                                                       | Entering a username during login.                                                           | To let the system know who is attempting to access it.                                                     |
| **Authentication**                | Verifying the claimed identity by checking credentials.                                          | Entering a password or using two-factor authentication.                                      | Ensures the user is who they say they are.                                                                 |
| **Authorization**                 | Determining which resources a user can access after authentication.                             | A shipping employee accessing logistics systems.                                              | Controls access to resources based on user roles or permissions.                                           |
| **Accounting**                    | Tracking and logging user activities on the system.                                              | Logging user login times, data usage, and logout times.                                      | Enables monitoring of user actions for security and auditing purposes.                                     |
| **AAA Framework**                 | Authentication, Authorization, and Accounting framework for secure access control.              | VPN concentrator using AAA to allow secure access to an internal file server.                | Provides a structured process for managing secure access to network resources.                             |
| **AAA Server**                    | A centralized server for managing user authentication and authorization.                        | VPN server communicating with a central AAA server to verify user credentials.               | Separates authentication from individual systems, making it more scalable and manageable.                   |
| **Certificate-Based Authentication** | Using digital certificates to verify the identity of devices.                                   | Company laptop using a certificate to authenticate with the VPN concentrator.                | Provides device authentication without storing passwords on field devices.                                 |
| **Certificate Authority (CA)**    | An entity that issues and signs digital certificates.                                            | A CA issuing a certificate for a device to prove its identity on the network.                | Ensures the authenticity of certificates and enables trust in the authentication process.                  |
| **Authorization Models**          | Frameworks used to assign permissions based on user roles, attributes, or other characteristics. | Role-Based Access Control (RBAC), Attribute-Based Access Control (ABAC).                     | Defines how users are authorized to access resources.                                                      |
| **Scalability in Authorization**  | The ability to efficiently manage access rights for many users by grouping them into roles.      | Grouping users in Shipping and Receiving to give access to resources without individual mapping. | Simplifies permission management for large organizations by avoiding manual assignment of rights.           |

### Scrumban
1. Kanban Board in Scrum
The Scrum team can use a Kanban board to visualize the workflow and track the
progress of work during each Sprint.

The board typically has columns for "To Do", "In Progress", and "Done", mirroring
the Scrum workflow.

Work items (user stories, tasks, bugs, etc.) are represented as cards on the board
and moved across the columns as they progress.


2. Limiting Work in Progress (WIP)
Scrum teams can apply WIP limits to each column on the Kanban board to
prevent overloading the team and ensure a steady flow of work.

This helps identify bottlenecks and encourages the team to focus on completing
work before starting new tasks.

3. Continuous Improvement
The Kanban board provides transparency into the workflow, allowing the Scrum
team to identify areas for improvement during the Sprint Retrospective.

Teams can experiment with adjusting WIP limits, changing workflow steps, or
other incremental process changes to optimize their Scrum process.


4. Complementary Practices

Scrum ceremonies like Daily Standup, Sprint Planning, and Sprint Review can be
conducted alongside the Kanban board visualization.

The Product Backlog and Sprint Backlog still form the foundation of the Scrum
process, with the Kanban board providing an additional layer of workflow
management
